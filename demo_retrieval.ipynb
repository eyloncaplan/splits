# Bare Bones BM25 Retrieval and Lift Evaluation

This notebook demonstrates how to load a Pyserini index, query it with a predefined lexicon (seed words), and compute the Lift metric for a target demographic.

```python
# 1. Import Core Logic Elements
from core_logic import (
    query_bm25_index, 
    lift_at_k, 
    lift_ci,
    compute_keyword_similarity
)
```

```python
# 2. Execution Example

# Set the path to the index
# Update the index name with whatever you are evaluating (e.g. 'Video_Games_-General_Discussion-_black-jewish')
index_name = 'Video_Games_-General_Discussion-_black-jewish'
index_path = f'/homes/ecaplan/gscratch1/bm25_indexes/gt_reranking_sets/{index_name}'

# Target demographic mapping
target_demo = 'jewish'

# Define seed words (or load from demo_seed_words.json)
keywords = [
    "the", "be", "to", "of", "and",
    "a", "in", "that", "have", "I",
    "it", "for", "not", "on", "with",
    "he", "as", "you", "do", "at"
]

# Run search (fetch top 1000 items)
df_results = query_bm25_index(index_path, keywords, doc_count=2000)

print(f"Total retrieved docs: {len(df_results)}")

# Compute lift at thresholds
lift_0_5_percent = lift_at_k(df_results, target_demo, k=0.05)
pval_0_5_percent, ci_lower_0_5_percent, ci_upper_0_5_percent = lift_ci(df_results, target_demo, k=0.05)

lift_1_percent = lift_at_k(df_results, target_demo, k=0.01)
pval_1_percent, ci_lower_1_percent, ci_upper_1_percent = lift_ci(df_results, target_demo, k=0.01)

lift_2_percent = lift_at_k(df_results, target_demo, k=0.02)
pval_2_percent, ci_lower_2_percent, ci_upper_2_percent = lift_ci(df_results, target_demo, k=0.02)

lift_5_percent = lift_at_k(df_results, target_demo, k=0.05)
pval_5_percent, ci_lower_5_percent, ci_upper_5_percent = lift_ci(df_results, target_demo, k=0.05)

lift_10_percent = lift_at_k(df_results, target_demo, k=0.10)
pval_10_percent, ci_lower_10_percent, ci_upper_10_percent = lift_ci(df_results, target_demo, k=0.10)

print(f"Lift@0.5%: {lift_0_5_percent:.3f} (p-value: {pval_0_5_percent:.4f}, 95% CI: [{ci_lower_0_5_percent:.3f}, {ci_upper_0_5_percent:.3f}])")
print(f"Lift@1%:  {lift_1_percent:.3f} (p-value: {pval_1_percent:.4f}, 95% CI: [{ci_lower_1_percent:.3f}, {ci_upper_1_percent:.3f}])")
print(f"Lift@2%:  {lift_2_percent:.3f} (p-value: {pval_2_percent:.4f}, 95% CI: [{ci_lower_2_percent:.3f}, {ci_upper_2_percent:.3f}])")
print(f"Lift@5%:  {lift_5_percent:.3f} (p-value: {pval_5_percent:.4f}, 95% CI: [{ci_lower_5_percent:.3f}, {ci_upper_5_percent:.3f}])")
print(f"Lift@10%: {lift_10_percent:.3f} (p-value: {pval_10_percent:.4f}, 95% CI: [{ci_lower_10_percent:.3f}, {ci_upper_10_percent:.3f}])")

# Display top results
df_results
```
